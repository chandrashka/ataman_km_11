```python
from math import sqrt
def triangle_ineq(func):
    def inner(a,b,c):
        if check(a,b,c):
            a,b,c = float(a), float(b), float(c)
            if a + b >= c and a+c>=b and c+b>=a:
                func(a,b,c)
            else:
                print("Ви ввели не трикутник :(")
        elif not check(a,b,c):
            print("От халепа, Ви ввели не числа")
    return inner

@triangle_ineq
def area_calculations(a,b,c):
    p = (a+b+c)/2
    print("Площа вашого трикутника:",sqrt(p*(p-a)*(p-b)*(p-c)))
def check(a,b,c):
    try:
        if float(a) > 0 and float(b) > 0 and float(c) > 0:
            return True
    except:
        return False
a,b,c = input("Введіть довжини трьох сторін трикутника через пробіл: ").split()
area_calculations(a,b,c)


```

```python

while 1:
    try:
        n = int(input("Enter your number: "))
        if n > 1280 or n % 16 != 0 or n == 0:
            print("Error")
            continue
        else:
            break
    except:
        print("Error")
def decor(arg):
    def inner(func):
        if arg == False:
            func(n)
        else:
            lst = []
            for i in func(n):
                lst.append(i)
            print("Всі сторірки книги:\n",lst)
    return inner
    
@decor(arg=True)
def zosh(n):
    zosh_num = n //16
    i = 0
    print("Кількість зошитів:", zosh_num)
    while i < zosh_num:
        i += 1
        end = i *16
        start = i * 16 - 15
        ark = []
        for j in range(4):
            ark.append(end)
            end -= 1
            ark.append(start)
            ark.append(start+1)
            start += 2
            ark.append(end)
            end -= 1

        print(i, "зошит має сторінки:\n",ark)
        yield ark
    
    
```
